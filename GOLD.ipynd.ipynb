```python
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

outliers_by_column = {}

for column in numeric_columns:
    z_scores_col = zscore(data[column])  
    outliers = data[np.abs(z_scores_col) > 4]  
    if not outliers.empty:
        outliers_by_column[column] = outliers[[column]]

for column, outliers in outliers_by_column.items():
    print(f"القيم الشاذة في العمود '{column}':")
    print(outliers)
    print("-" * 50)
```

```python
umeric_columns = data.select_dtypes(include=[np.number]).columns
z_scores = data[numeric_columns].apply(zscore)
threshold = 3
data_cleaned = data[(np.abs(z_scores) <= threshold).all(axis=1)]
print(data_cleaned.head())
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.set_index('Date', inplace=True)
```

```python
print(plt.style.available)

plt.style.use('seaborn-v0_8-dark-palette')
plt.figure(figsize=(14, 7))

plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price')
plt.title('Gold Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

```python
decomposition = seasonal_decompose(data['Close'], model='multiplicative', period=365)
decomposition.plot()
plt.show()
```

```python
train = data['Close'][:int(0.8 * len(data))]
test = data['Close'][int(0.8 * len(data)):]
```

```python
model = ARIMA(train, order=(5, 1, 0))
model_fit = model.fit()
residuals=model_fit.resid
```

```python
plt.Figure(figsize=(14,7))
plt.plot(residuals,marker='o',linestyle='-')
plt.axhline(y=0,color='r',linestyle='--')
plt.title('Residual plot')
plt.xlabel('Time periods')
plt.grid()
plt.show()
```

```python
forecast = model_fit.forecast(steps=len(test))
test.index = forecast.index 
```

```python
plt.figure(figsize=(14, 7))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f'RMSE: {rmse}')
```
